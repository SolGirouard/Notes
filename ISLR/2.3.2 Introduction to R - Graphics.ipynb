```r
x = seq(-pi, pi, length=50)
```

```r
seq(1,10,0.5)
```

```r
# Contour plot for 3D data
y=x
f=outer(x,y,function(x,y)cos(y)/(1+x^2))
contour(x,y,f)
```

```r
fa=(f-t(f))/2
contour(x,y,fa,nlevels=15)
```

```r
# Image works like contour but produces a heatmap
image(x,y,fa)
```

```r
#persp produces a 3D plote
persp(x,y,fa)
```

```r
persp(x,y,fa,theta=30)
```

```r
persp(x,y,fa,theta=30, phi=40)
```
